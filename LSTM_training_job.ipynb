## Training Job for LSTM EEG

```python
from sagemaker.pytorch import PyTorch
import sagemaker

role = sagemaker.get_execution_role()

BUCKET = "eeg-denoise"
PREFIX = "prepared"

estimator = PyTorch(
    entry_point='train.py',
    source_dir='src',
    role=role,
    framework_version='2.1.0',
    py_version='py310',
    instance_count=1,
    instance_type='ml.g4dn.xlarge',  # Instance type (e.g., GPU-enabled)
    max_wait=3600,               # Max total time (scheduling + training)
    max_run=1800,                # Max uninterrupted training time
    use_spot_instances=True,     # Spot mode ON
    checkpoint_s3_uri=f's3://{BUCKET}/{PREFIX}/checkpoints',
    hyperparameters={
        'epochs': 15
    },
    output_path=f's3://{BUCKET}/{PREFIX}/output',
    base_job_name='eeg-denoising-job',
)

estimator.fit({
    'train': f's3://{BUCKET}/{PREFIX}/train',
    'val': f's3://{BUCKET}/{PREFIX}/val',
})
```
